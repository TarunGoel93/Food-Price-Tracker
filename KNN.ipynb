```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\Dell\Downloads\Intership\Projects\Food Price Tracker\cleaned_WLD_RTFP_country_2023-10-02.xlsx"
df = pd.read_excel(file_path )
print(df)
```

```python
X = df[['year', 'Close']]
```

```python
x=df.iloc[:,-1]
y=df.iloc[:,3]
```

```python
x
```

```python
y
```

```python
k = 3   # cluster is defined 
```

```python
kmeans = KMeans(n_clusters = k , random_state = 42)
```

```python
# Fit the model
kmeans.fit(X)
```

```python
# get cluster centroids
centroids = kmeans.cluster_centers_
```

```python
# Labels of clusters
labels = kmeans.labels_
```

```python
print(centroids)
print(labels)
```

```python
plt.scatter(x,y, c=labels, cmap='viridis')
plt.scatter(centroids[:,0],centroids[:,1] , s=300 , c="red" , marker='o')
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.title('Food Price Tracker')
```
